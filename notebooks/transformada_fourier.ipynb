```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de la seal
Fs = 1000        # Frecuencia de muestreo (Hz)
T = 1            # Duracin de la seal en segundos
N = Fs * T       # Nmero de puntos de datos
t = np.linspace(0, T, N)  # Vector de tiempo
# Generar una seal compuesta por dos frecuencias sinusoidales
f1 = 10    # Frecuencia de la primera sinusoide (Hz)
f2 = 50    # Frecuencia de la segunda sinusoide (Hz)
f3 = 90    # Frecuencia de la segunda sinusoide (Hz)
f4 = 75    # Frecuencia de la segunda sinusoide (Hz)
A1 = 1.0   # Amplitud de la primera sinusoide
A2 = 0.5   # Amplitud de la segunda sinusoide
print(f'Samples before {t}')
tiempo=t.copy()
for index,sample in enumerate(t):
    if index<100:
        t[index]*=f3
    elif index>800:
        t[index]*=f4
    else:
        t[index]*=f1
print(f'Samples after {t}')
signal = A1 * np.sin(2 * np.pi * t) + A2 * np.sin(2 * np.pi * t)

# Calcular la transformada de Fourier de la seal
fourier_transform = np.fft.fft(signal)
frequencies = np.fft.fftfreq(N, 1 / Fs)  # Vector de frecuencias correspondiente
print(f'Frecuencias {frequencies} len {len(frequencies)}')
print(f'Fourier transform {fourier_transform} len {len(fourier_transform)}')
# Graficar la seal en el dominio del tiempo
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(tiempo, signal)
plt.title('Seal en el dominio del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

# Graficar la magnitud de la transformada de Fourier
plt.subplot(1, 2, 2)
plt.plot(frequencies[0:100], np.abs(fourier_transform[0:100]))
plt.title('Transformada de Fourier (Magnitud)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
#plt.xlim(0, Fs / 2)  # Mostrar solo la mitad positiva del espectro de frecuencia
plt.tight_layout()
plt.show()
```
